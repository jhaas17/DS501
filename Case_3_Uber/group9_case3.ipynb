# Problem 1: Predicting Distance

## Data Collection/Processing

```python
import pandas as pd 

data_jan = pd.read_csv("data/yellow_tripdata_2018-01.csv")
weather = pd.read_csv("data/central_park_weather.csv")

# Preview the first 5 lines of the loaded data 
data_jan.head()
```

```python
weather.head()
```

```python
weather_jan = weather[(weather['DATE'] >= '2018-01-01') & (weather['DATE'] <= '2018-01-31')]
weather_jan = weather_jan[['DATE','AWND','PRCP','SNOW','SNWD','TMAX','TMIN']]
weather_jan['DATE'] = pd.to_datetime(weather_jan['DATE'], format="%Y-%m-%d")
weather_jan['MonthDayNum'] = weather_jan['DATE'].dt.day
weather_jan.head(31)
```

```python
data_jan['Duration'] = (pd.to_datetime(data_jan['tpep_dropoff_datetime'], format="%Y-%m-%d %H:%M:%S") - pd.to_datetime(data_jan['tpep_pickup_datetime'], format="%Y-%m-%d %H:%M:%S")).astype(int)/10**9
data_jan.head()
```

```python
df_jan = data_jan[['tpep_pickup_datetime','Duration','trip_distance','passenger_count','fare_amount','extra','mta_tax','tip_amount','tolls_amount','total_amount']]
df_jan.head()
```

```python
df_jan.shape
```

```python
df_jan['tpep_pickup_datetime'] = pd.to_datetime(df_jan['tpep_pickup_datetime'], format="%Y-%m-%d %H:%M:%S")
df_jan['MonthDayNum'] = df_jan['tpep_pickup_datetime'].dt.day
df_jan['DayOfWeekNum'] = df_jan['tpep_pickup_datetime'].dt.dayofweek
df_jan['pickup_time'] = df_jan['tpep_pickup_datetime'].dt.time

final_df = pd.merge(df_jan, weather_jan, on='MonthDayNum')
final_df.head()
```

## The Solution: implement a data science solution to the problem you are trying to solve.

Merge weather and tripdata datasets using Date as Key from January 2018 to December 2018, calculate the duration and retrieve time details for each trip. we got 8759847 Trips in January 2018. The training feature would be trip distance, passenager count, total amount of the trip, and weather factors while the target feature would be trip duration. The trip duration estimation model is developed based on that. We are using data from January 2018 as training dataset and February 2018 as testing dataset. Finally we can customized the value in this model and come up with the estimated duration.

```python
x_train = []
x_train = final_df[['trip_distance','passenger_count','total_amount','DayOfWeekNum','AWND','PRCP','SNOW','TMAX','TMIN']].values.tolist()
y_target = final_df[['Duration']].values.tolist()
```

# Problem 2: Predicting Fares

## Data Collection/Processing: 

```python
# We are choosing to work with January and February data
import pandas as pd
jan = pd.read_csv('data/yellow_tripdata_2018-01.csv', usecols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'fare_amount', 'PULocationID'])
feb = pd.read_csv('data/yellow_tripdata_2018-02.csv', usecols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'fare_amount'])
# mar = pd.read_csv('yellow_tripdata_2018-03.csv', usecols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'fare_amount'])
# apr = pd.read_csv('yellow_tripdata_2018-04.csv', usecols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'fare_amount'])
# may = pd.read_csv('yellow_tripdata_2018-05.csv', usecols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'fare_amount'])
# jun = pd.read_csv('yellow_tripdata_2018-06.csv', usecols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_distance', 'fare_amount'])
# uber2018 = pd.concat([jan,feb,mar,apr,may,jun], ignore_index = True)
```

```python
def process_data(uber2018):
    uber2018.tpep_dropoff_datetime = pd.to_datetime(uber2018.tpep_dropoff_datetime, format='%Y-%m-%d %H:%M:%S')
    uber2018.tpep_pickup_datetime = pd.to_datetime(uber2018.tpep_pickup_datetime, format='%Y-%m-%d %H:%M:%S')
    uber2018['duration'] = uber2018['tpep_dropoff_datetime']-uber2018['tpep_pickup_datetime']
    uber2018['dayofweek'] = pd.Series(uber2018['tpep_pickup_datetime']).dt.dayofweek
    uber2018['hour'] = uber2018.tpep_pickup_datetime.apply(lambda x: x.hour)
    uber2018 = uber2018.loc[uber2018['trip_distance'] > 0]
    uber2018 = uber2018.loc[uber2018['fare_amount'] > 0]
    uber2018['duration'] = uber2018['duration'].astype(int)/10**9/60
    uber2018 = uber2018.loc[uber2018['duration'] > 0]
    uber2018['price_per_mile'] = uber2018.fare_amount/uber2018.trip_distance
    uber2018['price_per_min'] = uber2018['fare_amount']/uber2018['duration']
    return uber2018
```

```python
uber_jan = process_data(jan)
uber_jan.head()
```

```python
import matplotlib.pyplot as plt
# Plot price per mile by hour
avg_by_hour = uber_jan['price_per_mile'].groupby(uber_jan['hour']).mean()
avg_by_hour.plot(kind = 'bar')
plt.show()
# Plot price per minute by hour
avg_by_hour_dur = uber_jan['price_per_min'].groupby(uber_jan['hour']).mean()
avg_by_hour_dur.plot(kind = 'bar')
plt.show()
# Plot price per mile by day
avg_by_day = uber_jan['price_per_mile'].groupby(uber_jan['dayofweek']).mean()
avg_by_day.plot(kind = 'bar')
plt.show()
# Plot price per minute by day
avg_by_day_dur = uber_jan['price_per_min'].groupby(uber_jan['dayofweek']).mean()
avg_by_day_dur.plot(kind = 'bar')
plt.show()
```

```python
# Plot price per mile by day
avg_by_hour_fri = uber_jan.loc[uber_jan['dayofweek'] == 4]['price_per_mile'].groupby(uber_jan['hour']).mean()
avg_by_hour_fri.plot(kind = 'bar')
plt.show()
# Plot price per minute by day
avg_by_hour_fri_dur = uber_jan.loc[uber_jan['dayofweek'] == 4]['price_per_min'].groupby(uber_jan['hour']).mean()
avg_by_hour_fri_dur.plot(kind = 'bar')
plt.show()
```

## The Solution: implement a data science solution to the problem you are trying to solve.

We will fit a linear model using January data from 2018. We will consider day, hour, distance, and duration as predictors. We will then use the model to predict the cost of rides from the February data set. Finally, we will examine the accurcy of the predictions.

```python
train_data = uber_jan[['dayofweek','hour','trip_distance', 'duration']].values.tolist()
train_target = uber_jan[['fare_amount']].values.tolist()
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
model = model.fit(train_data, train_target)
```

```python
uber_feb = process_data(feb)
```

```python
test_data = []
test_data = uber_feb[['dayofweek','hour','trip_distance', 'duration']].values.tolist()
test_target = []
test_target = uber_feb[['fare_amount']].values.tolist()
```

```python
test_pred = model.predict(test_data)
```

## Results: summarize and visualize the results discovered from the analysis

```python
import numpy as np
import sklearn.metrics as metrics 

print(metrics.mean_absolute_error(test_target, test_pred)/len(test_pred))
print(np.sqrt(metrics.mean_squared_error(test_target, test_pred))/len(test_pred))
```

The results show that we are able to predict the price quite accurately. We achieve a mean absolute error of 7.6e-7 and a mean squared error of 1.2e-6. This high accuracy is likely due to how much taxi services use time and distance to calulcate price. Since time and distance (gas) are the two main costs of operating such a business, the cost must reflect these values.
